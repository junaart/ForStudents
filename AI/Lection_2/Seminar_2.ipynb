# Работа с пакетом neurolab

## Создание многослойного перцептрона

**neurolab.net.newff(minmax, size, transf=None)**

Параметры: 

minmax: список списков, внешний список — это количество входных нейронов, внутренние списки должны содержать 2 элемента: min и max - диапазон входных значений;

size: длина списка равна количеству слоев, кроме входного слоя, элемент списка — количество нейронов соответствующего слоя

transf: список активационных функций для каждого слоя (по умолчанию TanSig)

Построим нейронную сеть из одного входного нейрона, а также двух внутренних слоей по 5 и 3 нейрона в каждом и одного нейрона на выходе:

```python
import neurolab as nl
net = nl.net.newff([[0,10], [0,10], [0,10], [0, 10]], [5, 1])
```

```python
import neurolab as nl
import numpy as np

# Create train samples
x = np.linspace(-7, 7, 20)
y = np.sin(x) * 0.5

size = len(x)

inp = x.reshape(size,1)
tar = y.reshape(size,1)

# Create network with 2 layers and random initialized
net = nl.net.newff([[-7, 7]],[5, 10,100,1])

# Train network
error = net.train(inp, tar, epochs=500, show=100, goal=0.02)

# Simulate network
out = net.sim(inp)

# Plot result
import pylab as pl
pl.subplot(211)
pl.plot(error)
pl.xlabel('Epoch number')
pl.ylabel('error (default SSE)')

x2 = np.linspace(-6.0,6.0,150)
y2 = net.sim(x2.reshape(x2.size,1)).reshape(x2.size)

y3 = out.reshape(size)

pl.subplot(212)
pl.plot(x2, y2, '-',x , y, '.', x, y3, 'p')
pl.legend(['train target', 'net output'])
pl.show()
```
