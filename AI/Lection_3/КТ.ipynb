# Контрольная работа

По набору данных в файле 2.xls построить две модели:

1. По входным характеристикам: пол, возраст, стаж, оклад, семейное положение построить нейронную сеть решающую задачу классификации (есть дети, нет детей);

2. По входным характеристикам: пол, возраст, стаж, оклад, семейное положение построить нейронную сеть решающую задачу регрессии (сколько детей).

# Задача классификации 

```python
#Читаем данные
import pandas as pd
import numpy as np
from sklearn import preprocessing

D = pd.read_excel('2.xls')
D_n = pd.DataFrame({'Пол': preprocessing.scale(D['Пол'].rank()),
                    'Возраст': preprocessing.scale(D['Возраст']),
                    'Стаж': preprocessing.scale(D['Стаж']),
                    'Оклад': preprocessing.scale(D['Оклад']),
                    'Семья': preprocessing.scale(D['Семейное положение'].rank()),
                    'Дети': [0 if (np.isnan(i)) else i for i in D['Кол-во детей']]})
display(D_n)
```

```python
#Формируем обучвющую и тестовую выборки
from keras.utils import to_categorical
train_set = D_n.iloc[0:81, 0:5]
train_labels = to_categorical([i if i == 0 else 1 for i in list(D_n.iloc[0:81, 5:6]['Дети'])])
test_set = D_n.iloc[81:100, 0:5]
test_labels = to_categorical([i if i == 0 else 1 for i in list(D_n.iloc[81:100, 5:6]['Дети'])])
```

```python
#Конструируем сеть

from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(6, activation='relu', input_shape = (5,)))
network.add(layers.Dense(7, activation='relu'))
#network.add(layers.Dense(40, activation='relu'))
network.add(layers.Dense(2, activation='softmax'))

#Компиляция сети
network.compile(optimizer = 'rmsprop',
                loss = 'categorical_crossentropy',
                metrics=['accuracy'])
```

```python
#Обучаем сеть
res = network.fit(train_set,train_labels, epochs = 200, batch_size = 60)
```

```python
test_loss, test_acc = network.evaluate(test_set, test_labels)
print(test_loss)
print(test_acc)
```

```python
ynew = network.predict(test_set)
```

```python
import numpy as np

for i in range(len(test_labels)):
    print(np.argmax(test_labels[i]), np.argmax(ynew[i]))
```

```python
count_error = 0

for i in range(len(ynew)):
    if np.argmax(test_labels[i]) != np.argmax(ynew[i]):
        count_error += 1
print(count_error, len(ynew))
```
