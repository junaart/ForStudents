# Вопросы к экзамену по дисциплине "Управление и администрирование информационных систем"

## Теоретические вопросы

1. Определение Управления(администрирования) ИС. Группы задач администрирования.
2. Состав служб администратора системы и их назначение.
3. Определение Информационной системы (ИС). Компоненты ИС. Понятия Технических средств ИС, Программных и технологических средств ИС, Информационного фонда.
4. Определение Корпоративных ИС.  Классы корпоративных ИС: MRP, MRP II, ERP, ERP-II и CSRP.
5. Основы администрирования в операционной системе Linux: общие команды (примеры использования).
6. Структура файловой системы Linux.
7. Команды работы с файлами в Linux.
8. Назначение прав доступа к файлам в Linux.
9. Перенаправление ввода/вывода, исполььзование каналов, утилита grep.
10. Монтирование устройств, монтирование папок. Назначение каталолга /dev.
11. Понятие процесса, команды работы с процессами. Сигналы.
12. Архиватор tar, утилита gzip - назначение, примеры использования.
13. Организация резервного копирования с помощью rsync. Примеры использования.
14. Планирование заданий - демон cron.
15. Резервирование с помощью технологии git.
16. Командные оболочки (shell) - назначение, разновидности, история возникновения.
17. Операторы сравнения, условные конструкции в shell, примеры использования.
18. Циклические конструкции в shell, примеры использования.
19. Удаленное управление компьютером с помощью ssh.
20. Модель OSI, основные протоколы сети интернет.
21. Адрес IPv4. Типы. Формы записи адреса IPv4.
22. Классы IPv4 адресов. Их описание.
23. Понятие маски подсети. Формы записи маски подсети. Бесклассовая маршрутизация.
24. Публичные и приватные адреса IPv4. Зарезервированные IP-адреса.
25. Основные команды проверки и настройки сети.

## Примеры практических заданий

**Задание 1**.

Создать скрипт, который резервирует всю информацию из рабочего каталога в каталог Rezerve (данный каталог также разместить в рабочем каталоге), причем, если в каталоге Rezerve уже есть каталоги c аналогичными именами, то перед добавлением их необходимо рекурсивно удалить (каталог Rezerve внутрь самого себя добавлять не нужно). Сделать запись в cron для  запуска данного скрипта ежедневно в 12.00 AM.

**Задание 2**.

- В рабочем каталоге создайте создайте папку «Test_one», внутри которой еще две папки «Example1» «Example2».
- Установите на папку «Example1» права для владельца и группы владельца: чтение, запись, выполнение, SUID и SGID, для остальных права запись.
- Создайте в папке «Example1» три файла «Текст 1», «Текст 2», «Текст 3».
- Без использования текстового редактора запишите в файл «Текст 1» следующий текст: «Магистром можешь ты не быть! Но бакалавром быть обязан!».
- Без использования текстового редактора скопируете содержимое файла «Текст 1» в файл «Текст 2».
- Переместите файл «Текст 2» в папку «Example2» и скопировать файл «Текст 1» в папку «Example2».
- Сожмите файлы «Текст 1», «Текст 2» из папки «Example2» с помощью утилиты gzip и добавьте их в архив «MyAchiv» с помощью утилиты tar.
- Добавьте в файл «Текст 3» содержимое «Текст 1» без использования текстового редактора, а далее в конец «Текст 3» добавьте строку «Обязан быть магистром, коль этот пункт я одолел!»
- В папке «Test_one» создайте файл «Result» с результатами поиска слова, начинающего ся на «магистр», из файла «Текст 3».
- Добавьте, без использования текстового редактора, в конец файла «Текст 3» полную информацию о ядре системы и текущую дату.

**Задание 3**.

Создать скрипт, добавляющий в выбранную папку 10 файлов со случайным названием и случайно назначая права доступа к файлам его  владельцу. В качестве владельца файлов назначить student01.

**Задание 4**.

Создать скрипт, вкладывающий папку в папку заданное количество раз. Количество раз и имя папки ввести с клавиатуры.

**Задание 5**.

Настроить  удаленный доступ по протоколу ssh, удаленно создать скрипт, который резервирует информацию из заданной папки в другую папку, изменяя все права в резервной папке только на чтение для всех пользователей. Удаленно настроить планировщик заданий cron для запуска данного скрипта каждые 5 минут.
