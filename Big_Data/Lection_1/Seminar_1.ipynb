```python
class Node:
    __value = None
    __next = None

    def __init__(self,x):
        self.__value=x
        self.__next=None
        
    def set_value(self,x):
        self.__value = x

    def set_next(self,y):
        self.__next = y

    def get_value(self):
        return self.__value

    def get_next(self):
        return self.__next
```

```python
n1=Node(1)
n2=Node(2)
```

```python
n1.set_next(n2)
```

```python
n1.get_next()
```

```python
n2
```

```python
class mylist:
    __first = None
    __count = 0

    def add(self,x):
        a = Node(x)
        a.set_next(self.__first)
        self.__first = a
        self.__count += 1

    def get_count(self):
        return self.__count

    def get_first(self):
        return self.__first

    def print_list(self):
        x= self.__first
        while x.get_next() != None:
            print(x.get_value())
            x = x.get_next()
        print(x.get_value())
```

```python
#Создание хранилища ключ-значение
import uuid
class KeyValueStorage():
    __dic={}
    def add_record(self,keys,values):
        uuidm=uuid.uuid4().int
        f={}
        for i in range(len(keys)):
            f[keys[i]]=values[i]
        self.__dic[uuidm]=f
    def add_list(self, L_keys, L_values):
        for i in L_values:
            self.add_record(L_keys,i) 
    def get_dict(self):
        return self.__dic
    def __init__(self):
        self.__dic={} 
    def get_by_uuid(self,x_uuid):
        try:
            r=self.get_dict()[x_uuid]
            return r
        except Exception:
            return None              
```
