```python
#Изменим название столбца
data=data.rename(columns={'region': 'Регион'})
data
```

```python
#Получить только показатели
data[['Субсидии на ГЗ из местных бюджетов 2015','Субсидии на ГЗ из федерального бюджета 2015']]
```

```python
#Отберем все показатели  из двух регионов
data[(data['Регион']=='Республика Алтай') | (data['Регион']=='Республика Крым')]
```

```python
#Отберем  гранты из местного и федерального бюджета из Крыма и Алтая
data[(data['Регион']=='Республика Алтай') | #
     (data['Регион']=='Республика Крым')]['Гранты из местного бюджета 2015']+data[(data['Регион']=='Республика Алтай') | (data['Регион']=='Республика Крым')]['Гранты из федерального бюджета 2015']
```

```python
#Отберем все регионы, у которых показатели "Субсидии на ГЗ из федерального бюджета 2015", 
#"Субсидии на НИОКР из федерального бюджета 2015" больше 0
data[(data['Субсидии на ГЗ из федерального бюджета 2015']>0) & #
     (data['Субсидии на НИОКР из федерального бюджета 2015']>0)]['Регион']
```

```python
#Рассчитаем их количество
data[(data['Субсидии на ГЗ из федерального бюджета 2015']>0) #
     & (data['Субсидии на НИОКР из федерального бюджета 2015']>0)]['Регион'].count()
```
