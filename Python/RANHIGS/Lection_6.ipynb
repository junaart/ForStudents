## **$\S 1.$ Кортежи**

Кортеж (**tuple**)  - это **неизменяемая** упорядоченная коллекция элементов. Для определения кортежа используется перечисление его элементов в круглых скобках, или явное задание с использованием ключевого слова **tuple**:

```python
a = (1,2,3,4,4,5)
b = tuple([1,2,5,6,7,8,7])
```

С кортежем можно выполнять все те же действия, что и со списком, но только если они не приводят к его изменению, в противном случае будет выдано сообщение об ошибке. 

```python
dir(a)
```

В целом кортежи — это легковесные и безопасные структуры данных, которые следует использовать, когда коллекция не должна меняться после создания.

### **Основные операции с кортежами**

```python
t = (10, 20, 30, 40, 50)

# Доступ по индексу
print(t[0])    # 10
print(t[-1])   # 50

# Срезы
print(t[1:4])  # (20, 30, 40)

# Конкатенация
t2 = (60, 70)
print(t + t2)  # (10, 20, 30, 40, 50, 60, 70)

# Повторение
print(t2 * 2)  # (60, 70, 60, 70)

# Проверка наличия
print(20 in t)    # True
print(100 in t)   # False

# Длина
print(len(t))  # 5

# Методы
print(t.index(30))  # 2 (индекс первого вхождения)
print(t.count(20))  # 1 (количество вхождений)
```

Кортеж также часто используется в качестве распаковки значений в переменные:

```python
# Распаковка в переменные
point = (10, 20)
x, y = point
print(x, y)  # 10 20

# С распаковкой переменного количества элементов
a, *rest = (1, 2, 3, 4)
print(a)     # 1
print(rest)  # [2, 3, 4]
```

### **Примеры использования словарей**

Cловари незаменимы для быстрого доступа, группировки и структурирования данных с уникальными ключами. Рассмотрим несколько примеров.

```python
# Перевод баллов

dict_balls = {
    90: "A",
    80: "B", 
    70: "C",
    60: "D",
    0: "F"
}

def conv_grade(balls):
    res = 'F'
    for i in sorted(dict_balls.items()):
        if balls >= i[0]:
            res = i[1]
    return res
```

```python
conv_grade(98)
```

```python
#Подсчет количества элементов (гистограмма)

def count_symbols(L):
    res = {}
    for i in L:
        res.setdefault(i,0)
    for i in L:
            res[i] += 1
    return res        
```

```python
count_symbols('Папа у Маши силен в математике. Папа решает, а Маша сдает.')
```

```python
#Подсчет количества слов

def count_words(L):
    LL = L.split(' ')
    res = {}
    for i in LL:
        if i != '':
            res.setdefault(i,0)
    for i in LL:
        if i != '':
            res[i] += 1
    return res   
```

Имеются также и некоторые другие методы:

- Метод **difference_update()** используется для удаления из исходного множества всех элементов, которые присутствуют в другом итерируемом объекте (или нескольких объектах). Он изменяет исходное множество "на месте", не создавая новое.

- Метод **intersection_update()** используется для оставления в исходном множестве только тех элементов, которые присутствуют во всех указанных итерируемых объектах. Он изменяет исходное множество "на месте", удаляя все элементы, не соответствующие пересечению.

- Метод **symmetric_difference_update()** используется для обновления множества элементами, которые присутствуют только в одном из двух множеств (исключая общие элементы). Он изменяет исходное множество "на месте", оставляя в нем симметрическую разность.
