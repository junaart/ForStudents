# **Лекция 7. Обработка исключений**

## **$\S 1.$ Базовый синтаксис**

**Исключение (exception)** — это событие, возникающее во время выполнения программы, которое нарушает нормальный поток инструкций. В Python исключения представляют собой объекты, которые "выбрасываются" (raise) при возникновении ошибки.

Природа исключений:

- Исключения — это не ошибки синтаксиса (они обнаруживаются на этапе выполнения)

- Исключения могут быть предсказуемыми (деление на ноль) и непредсказуемыми (отсутствие файла)

Python следует принципу EAFP (Easier to Ask for Forgiveness than Permission — проще попросить прощения, чем разрешения). Суть принципа в том, что вместо того чтобы проверять (ask for permission), можно сразу делать и в случае ошибки попросить прощения (handle exception).

````python
import os
# LBYL (Look Before You Leap)
if os.path.exists("file.txt"):
    file = open("file.txt")
else:
    print("Файл не найден")
````

````python
# EAFP - стиль Python
try:
    file = open("file.txt")
except FileNotFoundError:
    print("Файл не найден")
````

Из представленного примера видно, что для обработки исключений используются специальные зарезервированные слова: **try**, **except**. Фактически мы пишем - попытайся (try) открыть файл, если возникнет ошибка чтения файла FileNotFoundError, то выведи сообщение, что файл не найден.

В последнем примере видно, что перестановка обработчика исключения не меняет результата. Так будет не всегда. Переставлять можно обработчики, которые находятся на одном уровне иерархии исключений.

Если в обработчике анализируются некоторые частные исключения и в случае, если ни одно из них не подошло, то должен запускаться обработчик всего класса таких исключений, то перемешивать их нельзя.
Для такого случая нужно придерживаться правила: **обработчики исключений должны располагаться от более специфичных (дочерних) к более общим (родительским)**.

````python
# Правильно: от частного к общему
try:
    pass #Что-то делаем
except FileNotFoundError:      # Специфичный
    print("Файл не найден")
except PermissionError:         # Специфичный
    print("Нет прав доступа")
except OSError:                 # Общий (родитель)
    print("Системная ошибка")
except Exception:               # Ещё более общий
    print("Любая ошибка")
````

````python
# Неправильно: общий обработчик раньше специфичного
try:
    pass #Что-то делаем
except OSError:                 # Общий - перехватит всё!
    print("Системная ошибка")
except FileNotFoundError:       # Никогда не выполнится
    print("Файл не найден")
````

Python обрабатывает исключения последовательно сверху вниз и выполняет первый подходящий обработчик. Если общий обработчик стоит раньше специфичного, он "перехватит" исключение, и специфичный никогда не сработает.

Самая полная форма обработки исключений в Python **try-except-else-finally**:

```
try:
    # Блок кода, который может вызвать исключение
    # Защищаемая операция
    
except ExceptionType1:
    # Обработчик для ExceptionType1
    
except ExceptionType2 as variable:
    # Обработчик для ExceptionType2 с доступом к объекту исключения
    
except (ExceptionType3, ExceptionType4):
    # Один обработчик для нескольких типов
    
except:
    # "Голый" except - перехватывает всё (НЕ РЕКОМЕНДУЕТСЯ)
    
else:
    # Выполняется ТОЛЬКО если try не вызвал исключений
    # Код, который должен выполниться при успешном выполнении try
    
finally:
    # Выполняется ВСЕГДА, независимо от возникновения исключения
    # Используется для освобождения ресурсов, закрытия соединений
```

Ниже представлен пример, демонстрирующий использование этих конструкций

````python
def divide_numbers(a, b):
    print(f"Пытаюсь разделить {a} на {b}")
    
    try:
        # БЛОК TRY - защищаемый код
        print("  Вход в блок try")
        result = a / b
        print(f"  Результат вычисления: {result}")
        
    except ZeroDivisionError as e:
        # БЛОК EXCEPT - обработка деления на ноль
        print(f"  EXCEPT: Деление на ноль - {e}")
        return "Ошибка: деление на ноль"
        
    except TypeError as e:
        # БЛОК EXCEPT - обработка неверных типов
        print(f"  EXCEPT: Неверный тип данных - {e}")
        return "Ошибка: неверный тип данных"
        
    except Exception as e:
        # БЛОК EXCEPT - обработка любых других исключений
        print(f"  EXCEPT: Неожиданная ошибка - {type(e).__name__}")
        return f"Ошибка: {e}"
        
    else:
        # БЛОК ELSE - выполняется только при успехе
        print("  ELSE: Блок успешного выполнения")
        print(f"  ELSE: Результат умножен на 2 = {result * 2}")
        return result
        
    finally:
        # БЛОК FINALLY - выполняется всегда
        print("  FINALLY: Очистка ресурсов")
        print("  FINALLY: Закрытие соединений и т.д.")

# Тестирование
print("=== Тест 1: Успешное деление ===")
result = divide_numbers(10, 2)
print(f"Результат: {result}\n")

print("=== Тест 2: Деление на ноль ===")
result = divide_numbers(10, 0)
print(f"Результат: {result}\n")

print("=== Тест 3: Неверный тип ===")
result = divide_numbers(10, "2")
print(f"Результат: {result}\n")
````

Полная форма try-except-else-finally обеспечивает:

- **Try** - защиту опасного кода

- **Except** - обработку конкретных ошибок

- **Else** - отделение успешного кода от защищаемого

- **Finally** - гарантированную очистку ресурсов

Особенности finally:

- Выполняется всегда, даже если в try или except есть return, break, continue

- Выполняется даже если в try возникло необработанное исключение

- Выполняется до того, как исключение будет проброшено дальше

- Идеален для освобождения ресурсов (файлы, соединения, блокировки)

**Важные нюансы**

````python
# FINALLY с return
def example_finally_return():
    try:
        return "TRY return"
    finally:
        return "FINALLY return"  # Переопределяет return из try!

print(example_finally_return())  # Выведет: "FINALLY return"

# Правильно: не используйте return в finally
def example_good():
    try:
        return "TRY return"
    finally:
        print("Очистка")  # Только действия, без return
````

````python
# ELSE для предотвращения перехвата

# ПЛОХО: else не используется
try:
    file = open("data.txt")
    data = file.read()
    # Если здесь ошибка, она будет перехвачена except
    process(data)  # Может вызвать ValueError
except FileNotFoundError:
    print("Файл не найден")
except ValueError:
    print("Ошибка обработки данных")  # А должна быть в else!

# ХОРОШО: используется else
try:
    file = open("data.txt")
except FileNotFoundError:
    print("Файл не найден")
else:
    # Здесь process не защищён, его ошибки не перехватываются
    data = file.read()
    process(data)  # ValueError не будет перехвачен (это правильно)
````

````python
# Вложенные try-except

def nested_exceptions():
    try:
        # Внешний try
        print("Внешний try")
        try:
            # Внутренний try
            print("  Внутренний try")
            raise ValueError("Внутренняя ошибка")
        except ValueError as e:
            print(f"  Внутренний except: {e}")
            raise  # Проброс во внешний уровень
        finally:
            print("  Внутренний finally")
    except ValueError as e:
        print(f"Внешний except: {e}")
    finally:
        print("Внешний finally")
````

**Списание средств со счета**

````python
def withdraw_money(balance, amount):
    """
    Снятие денег со счета
    """
    print(f"\n--- Снятие {amount} руб. со счета (баланс: {balance}) ---")
    
    try:
        # TRY: Пытаемся выполнить операцию
        if amount <= 0:
            raise ValueError("Сумма снятия должна быть положительной")
        
        if amount > balance:
            raise ValueError(f"Недостаточно средств. Доступно: {balance}")
        
        new_balance = balance - amount
        
    except ValueError as e:
        # EXCEPT: Обработка ошибок
        print(f"Ошибка: {e}")
        return balance  # Баланс не изменился
        
    except TypeError as e:
        # EXCEPT: Обработка ошибок
        print(f"Ошибка: {e}")
        return balance  # Баланс не изменился
        
    else:
        # ELSE: Выполняется только при успехе
        print(f"Успешно снято {amount} руб.")
        print(f"Новый баланс: {new_balance} руб.")
        return new_balance
        
    finally:
        # FINALLY: Логирование операции
        print(f"Операция снятия завершена. Время: {__import__('datetime').datetime.now()}")
````

**Конвертация валют с проверкой курса**

````python
def convert_currency(amount, from_currency, to_currency, exchange_rates):
    """
    Конвертация валюты с проверкой курса
    """
    print(f"\n Конвертация {amount} {from_currency} → {to_currency}")
    
    rate = None
    try:
        # TRY: Получаем курс и выполняем конвертацию
        if amount <= 0:
            raise ValueError("Сумма должна быть больше нуля")
        
        if not isinstance(amount, (int, float)):
            raise TypeError("Сумма должна быть числом")
        
        # Получаем курс
        rate_key = f"{from_currency}_{to_currency}"
        rate = exchange_rates.get(rate_key)
        
        if rate is None:
            raise KeyError(f"Курс {from_currency} -> {to_currency} не найден")
        
        if rate <= 0:
            raise ValueError(f"Некорректный курс: {rate}")
        
        converted = amount * rate
        
    except ValueError as e:
        print(f"Ошибка значения: {e}")
        return None
        
    except TypeError as e:
        print(f"Ошибка типа: {e}")
        return None
        
    except KeyError as e:
        print(f"Курс не найден: {e}")
        return None
        
    else:
        # ELSE: Успешная конвертация
        print(f"Курс: 1 {from_currency} = {rate} {to_currency}")
        print(f"Получено: {converted:.2f} {to_currency}")
        return converted
        
    finally:
        # FINALLY: Всегда выводим информацию об операции
        print(f"Текущие курсы: {exchange_rates}")
````

**Расчет процентов по вкладу с обработкой ошибок**

````python
def calculate_interest(principal, rate, years, compound_frequency=12):
    """
    Расчет сложных процентов по вкладу
    
    principal: сумма вклада
    rate: годовая процентная ставка (%)
    years: срок в годах
    compound_frequency: периодичность капитализации (месяцев)
    """
    print(f"\nРАСЧЁТ ПРОЦЕНТОВ ПО ВКЛАДУ")
    print(f"Сумма: {principal} руб.")
    print(f"Ставка: {rate}% годовых")
    print(f"Срок: {years} лет")
    print(f"Капитализация: раз в {compound_frequency} мес.")
    
    try:
        # TRY: Валидация и расчет
        print("\n1. Проверка входных данных...")
        
        if not isinstance(principal, (int, float)):
            raise TypeError(f"Сумма должна быть числом")
        
        if principal <= 0:
            raise ValueError(f"Сумма вклада должна быть положительной, получено {principal}")
        
        if rate < 0:
            raise ValueError(f"Ставка не может быть отрицательной, получено {rate}")
        
        if rate > 100:
            raise ValueError(f"Слишком высокая ставка: {rate}% (макс. 100%)")
        
        if years <= 0:
            raise ValueError(f"Срок должен быть положительным, получено {years}")
        
        if not isinstance(compound_frequency, int):
            raise TypeError(f"Периодичность должна быть целым числом")
        
        if compound_frequency not in [1, 3, 6, 12]:
            raise ValueError(f"Некорректная периодичность: {compound_frequency} (допустимо: 1,3,6,12)")
        
        print("2. Выполнение расчета...")
        
        # Формула сложных процентов
        periods_per_year = 12 / compound_frequency
        total_periods = periods_per_year * years
        rate_per_period = (rate / 100) / periods_per_year
        
        final_amount = principal * (1 + rate_per_period) ** total_periods
        total_interest = final_amount - principal
        
    except (TypeError, ValueError) as e:
        # EXCEPT: Ошибки типов и значений
        print(f"Ошибка в данных: {e}")
        return None, None
        
    except OverflowError as e:
        # EXCEPT: Переполнение при больших числах
        print(f"Результат слишком велик: {e}")
        return None, None
        
    else:
        # ELSE: Успешный расчет
        print("\nРАСЧЁТ ВЫПОЛНЕН УСПЕШНО!")
        print(f"Конечная сумма: {final_amount:,.2f} руб.")
        print(f"Начислено процентов: {total_interest:,.2f} руб.")
        print(f"Эффективная ставка: {((final_amount/principal)**(1/years)-1)*100:.2f}%")
        
        # Дополнительный анализ
        if total_interest > principal:
            print(f"Прибыль превышает сумму вклада!")
        elif total_interest > 0:
            print(f"Вклад принесёт прибыль")
        else:
            print(f"Вклад не принесёт прибыли")
            
        return final_amount, total_interest
        
    finally:
        # FINALLY: Всегда показываем сводку
        print(f"\nСВОДКА:")
        print(f" Начальная сумма: {principal} руб.")
        print(f" Годовая ставка: {rate}%")
        print(f" Срок: {years} лет")
        print(f" Капитализация: {compound_frequency} мес.")
````

````python
final, interest = calculate_interest(100000, 8.5, 3, 12)
````
