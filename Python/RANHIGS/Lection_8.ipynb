## **$\S 1.$ Понятие библиотеки в Python**

Библиотеки, несомненно, наиболее сильная сторона Python и одна из главных причин его популярности. Буквально для любой задачи в какой-либо библиотеке можно найти решение, причем, достаточно качественное, и ещё чаще профессиональное.

Сам термин **библиотека** является в некотором смысле собирательным. Под ним понимают любой из двух следующих терминов:
- **модуль** - это один файл с расширением .py, он содержит определенные пользователем функции, переменные, классы и может быть непосредственно импортирован (подключен) в программу;
- **пакет** - это каталог, содержащий несколько модулей и обязательно файл ```__```init```__```.py (в Python 3.11+ он не всегда обязателен, но традиционно используется).

Таким образом, пакет более масштабная конструкция.

Например, построим наш собственный модуль:

````python
#Мой собственный модуль (демонстрация)

My_var = 777

def my_name(name):
    '''Документация функции: выводит name пользователя'''
    print(f'Меня зовут {name}')
    return True
````

Сохраним данный файл с расширением .py и положим рядом с файлом программы, в которой мы будем использовать этот модуль.

Теперь его мы можем подключить к своей программе и использовать определенные в модуле функции и переменные:

````python
import my_module
````

````python
my_module.my_name('Артамонов Юрий Николаевич')
````

````python
my_module.My_var
````

Разберемся теперь с понятием **пакета**.

Структура пакета выглядит следующим образом:

```
mypackage/              # <- это пакет (директория)
├── __init__.py         # обязателен (может быть пустым)
├── module1.py
├── module2.py
└── subpackage/         # вложенный пакет
    ├── __init__.py
    └── module3.py
```

| **NumPy** | Фундаментальные числовые вычисления | Основа для почти всех остальных библиотек. Работа с многомерными массивами и математическими функциями на языке C, что обеспечивает высокую скорость |

| **Polars** | Высокопроизводительная работа с таблицами | **Главный вызов Pandas в 2026 году**. Написан на Rust, что делает его в **5-30 раз быстрее** для больших наборов данных. Сохраняет похожий, но более строгий и предсказуемый API |
| **DuckDB** | Встраиваемая аналитика через SQL | Позволяет выполнять сложные SQL-запросы прямо к CSV, Parquet или Pandas DataFrame. Отлично подходит для интерактивного анализа крупных файлов, не помещающихся в оперативную память |
