```python
class human():
    __password = '123'

    def get_birthday(self):
        return self.__birthday

    def get_gender(self):
        return self.__gender

    def get_name(self):
        return self.__name

    def login(self,x):
        return x == self.__password
            
    def set_name(self,x):
        p = input('Введите пароль:')
        if self.login(p):
            if isinstance(x, str):
                self.__name = x
            else:
                print('Ожидается строка на входе')
        else:
            print('Вы не сотрудник ЗАГС')
        
    def __init__(self,name, birthday, gender):
        if isinstance(name, str):
            self.__name = name
        if isinstance(birthday, str):
            self.__birthday = birthday
        if (gender == 'м') or (gender == 'ж'):
            self.__gender = gender
```
