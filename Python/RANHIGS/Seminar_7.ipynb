```python
def convert(text):
    vowels = "аеёиоуыэюяАЕЁИОУЫЭЮЯ"
    result = ""
    for vowel in text:
        if vowel not in vowels:
            result += vowel
    return result
```

**Задача 6.**
Тарабарский язык: реализовать функцию, которая принимает на вход предложение и преобразует его следующим образом: после каждой гласной буквы добавляется буква "c" и эта гласная. Например, слово "Эквилибристика" преобразуется к виду "Эсэквисилисибрисистисикаса".

```python
def remove_vowels(text):
    vowels = "аеёиоуыэюяАЕЁИОУЫЭЮЯaeiouAEIOU"
    result = ""
    for vowel in text:
        if vowel in vowels:
            result += vowel + "c" + vowel
        else:
            result += vowel
    return result



user_input = input("Введите предложение для проверки: ")
processed_text = remove_vowels(user_input)  
print(f"Текст на тарабарском:{processed_text}")
```

**Задача 7.** Для заданного $N$ реализовать функцию, которая вычисляет следующее выражение:
$$S = 1-2+3-4+5-6+7-\ldots \pm N$$

```python
def sum(N):
    total = 0
    for i in range(1, N + 1):
        if i % 2 == 1:      
            total = total + i
        else:              
            total = total - i
    return total

N = int(input("Введите N: "))
print(f"Сумма ряда = {sum(N)}")
```

```python
def calculate_sum(N):
    total_sum = 0
    for i in range(1, N + 1):
        if i % 6 in (2, 3, 4):
            total_sum -= i
        else:
            total_sum += i     
    return total_sum
    
n = int(input("Введите число N: "))
result = calculate_sum(n)
print("Результат (через цикл):", result)
```

**Задача 8.**

Реализовать функцию, которая принимает целое число и находит сумму его цифр.

```python
def sum_of_digits(number):
    number = abs(number)
    total_sum = 0
    while number > 0:
        total_sum += number % 10
        number //= 10     
    return total_sum
    
n = int(input("Введите целое число: "))
print("Сумма цифр (через математику):", sum_of_digits(n))
```

```python
def sum_digits(N):
    res = 0
    for i in tuple(str(N)):
        res += int(i)
    return res
```

```python
portfolio = {
    "AAPL": {"shares": 100, "buy_price": 175.00},
    "GOOGL": {"shares": 50, "buy_price": 135.20},
    "MSFT": {"shares": 75, "buy_price": 330.50},
    "TSLA": {"shares": 30, "buy_price": 240.30},
    "AMZN": {"shares": 25, "buy_price": 145.80}
}

current_prices = {
    "AAPL": 178.50,
    "GOOGL": 140.15,
    "MSFT": 325.00,
    "TSLA": 235.50,
    "AMZN": 150.25
}
```

```python
r = [] #создание пустого списка
for t, d in portfolio.items():
    buy_price = d["buy_price"]
    current_price = current_prices[t]
    return_n = ((current_price - buy_price) / buy_price) * 100
    r.append((t, round(return_n, 4))) #двойные скобки, поскольку кортеж
print(r)
```

```python
def price_buy(act):
    if portfolio.get(act):
        return portfolio[act]["shares"] * portfolio[act]["buy_price"]

def profit(act):
    if portfolio.get(act):
        return current_prices[act] * portfolio[act]["shares"] - price_buy(act)

res = {}
for i in current_prices:
    res[i] = profit(i)

from operator import itemgetter
sorted_items = sorted(res.items(), key=itemgetter(1),reverse = True)
sorted_items
```
