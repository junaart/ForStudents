## Типы структурированных данных

 1. Непрерывные данные (continuous)}: данные, которые могут принимать любое значение в интервале. Так, любые числа из интервала $(0,1)$, например, 0.5674, являются непрерывными данными. Синонимы: интервал, число с плавающей точкой, числовое значение.
 
2. Дискретные данные (discrete)}: данные, которые могут принимать только целочисленные значения. Например, числа $\{-3, -2, -1, 0, 1, 2, 3, 4, 5\}$ являются дискретным набором данных. Синонимы: целое число, количество.

3. Категориальные (атрибутивные) данные (categorial)}: данные, которые могут принимать только определенные набор значений, в частности набор возможных категорий. Например, название городов, регионов - это категориальные данные. Синонимы: перечисления, факторы. В R факторы — это специальные переменные, которые  хранят данные в уровнях. Категориальная переменная — это те переменные, которые принимают значения на основе меток или имен. Например, группа крови человека может быть A, B, AB или O.

4. Двоичные (альтернативные) данные (binary): особый случай категориальных данных всего с двумя категориями значений (0/1, истина/ложь). Например, мужчина/женщина; брак/годен. Синонимы: дихотомические, логические, булевы данные.

5. Порядковые данные (ordinal)}: категориальные данные, которые допускают естественнное упорядочение. Например, набор воинских званий (лейтенант, старший лейтенант, капитан, майор, подполковник, полковник), (токарь 1 разряда, токарь 2 разряда, ... токарь 6 разряда). Синонимы: порядковый фактор.

```r
#Работа с непрерывными данными

#Получим вектор чисел с заданным законом распределения

n<-30
#Равномерный закон распределения runif(n, min=0, max=1)

x<-runif(n,1,10)
print(x)
print(mean(x))
print(11/2)
print(sd(x))
print(9/(2*sqrt(3)))
```

```r
#Нормальный закон распределения rnorm(n, mean=0, sd=1)

x<-rnorm(n,5,10)
print(x)
print(mean(x))
print(sd(x))
plot(x, xlab="Номер",ylab="Значение",main="График",type='b')
```

### Преобразование шкал

В ряде задач анализа данных требуется преобразовать значения показателей к единой безразмерной соизмеримой шкале. 
Для этих целей используются различные методы нормирования. 

Одним из наиболее популярных является так называемая **стандартизация данных**. 

Пусть имеются данные $x_1, x_2, \ldots, x_n$, тогда стандартизированными являются данные вида:
$$x^*_i=\frac{x_i-\overline{x}}{\sigma}$$
где $\overline{x}$ - средняя арифметическая, $\sigma$ - среднее квадратическое отклонение данных $x_1, x_2, \ldots, x_n$.

Покажем, что средняя арифметическая нормированных данных равна нулю:
   $$\overline{x^*}=\frac{\sum_{i=1}^{n}x^*_i}{n}=\frac{\sum_{i=1}^{n}\frac{x_i-\overline{x}}{\sigma}}{n}=\frac{1}{n\sigma}\left(\sum_{i=1}^{n}x_i-\sum_{i=1}^{n}\overline{x}\right)=\frac{1}{n\sigma}(n\overline{x}-n\overline{x})=0$$
   
Также покажем, что дисперсия стандартизированных данных равна единице:
     $$\frac{1}{n}\sum_{i=1}^{n}(x^*_i-\overline{x^*})^2=\frac{1}{n}\sum_{i=1}^{n}\left(\frac{x_i-\overline{x}}{\sigma}\right)^2=\frac{1}{\sigma^2}\cdot \frac{\sum_{i=1}^{n}(x_i-\overline{x})^2}{n}=\frac{\sigma^2}{\sigma^2}=1$$
     
Процедура стандартизации является обязательным этапом во многих многомерных методах статистического анализа: факторный, компонентный, кластерный анализ.

Кроме рассмотренного способа, одним из методов нормирования можно считать и переход от значений к их рангам.

На практике также могут встречаться случаи разнонаправленного влияния переменных, которые требуется привести к шкале $[0,N]$
1. Чем больше значение переменной, тем лучше:
$$x^*_i=\frac{x_i-x_{min}}{x_{max}-x_{min}}\cdot N$$
2. Чем меньше значение переменной, тем лучше:
$$x^*_i=\frac{x_{max}-x_i}{x_{max}-x_{min}}\cdot N$$
3. Чем ближе к $x_{\text{опт}}$, тем лучше:
$$x^*_i=\left(1-\frac{|x_i-x_{\text{опт}}|}{max\{(x_{max}-x_{\text{опт}}), (x_{\text{опт}}-x_{min})\}}\right)\cdot N$$

## Работа со структурированными данными в среде R: создание данных, загрузка, сохранение, чтение, удаление данных.

Одной из самых распространенных структур данных при анализе больших данных является датафрейм (датасет) - data.frame. По сути это простая плоская таблица, которая может иметь заголовки столбцов и заголовки строк, а также обеспечивает удобную навигацию по сгруппированным таким образом данным.

В языке R для этих целей используется data.frame:

```r
help(data.frame)
```

Датафрейм можно построить из отдельных векторов.

```r
#Введем несколько векторов даных
subject<-c('Кофта', 'Пиджак', 'Брюки','Юбка', 'Рубашка', 'Ботинки')
count<-c(5,7,8,3,2,6)

#Используем созданный ранее вектор size, и созданные сейчас вектора для создания фрейма
Data<-data.frame(size,subject, count)

#Присвоим имена столбцам данных
colnames(Data)<-c('size', 'subject','count')

#Присвоим имена строкам
rownames(Data)<-c('первый','второй','третий','четвертый','пятый','шестой')

#Посмотрим на созданный датафрейм
View(Data)

#По  датафрейму довольно просто строить различные диаграммы
pie(Data$count,labels=Data$size)

barplot(Data$count, names.arg=Data$size, border = "red", col="#33b7f2")

barplot(Data$count, names.arg=Data$size, border = "red", col="#33b7f2", horiz=TRUE)
```

Датафреймы можно не только создавать самостоятельно, но и загрузать из различных источников. Довольно популярным является формат csv.

```r
#Чтение данных в формате csv

people<-read.csv(file="demo.csv",sep=";",header=TRUE, fileEncoding = "UTF-8")

View(people)
```

## Манипулирование данными, обработка прямоугольных данных (таблиц)

К данным фрейма можно обращаться разными способами.

```r
#Получим столбец по его имени

print(Data$size)

#Получим строку по имени

print(Data["первый",])

#Получим столбец по индексу

print(Data[,2])

#Получим только заданные строки из заданных столбцов

print(Data[c("первый","второй"),c(1,2)])

#Получим заданный элемент строки и столбца

View(Data)

print(Data[2,3])

#Просуммируем разные элементы столбца

p<-Data$count[c(1:3)]+Data$count[c(4:6)]
print(p)

#Добавим новый столбец значений к датафрейму и дадим ему имя

Data<-data.frame(Data,sample(c(1:100),size=6))
colnames(Data)<-c(colnames(Data)[c(1:3)],"new")
View(Data)
```

```r
#Добавим новую строку к датафрему и присвоим ей имя

new_row<-c('XXL', "Пиджак", 8, 97)
Data<-rbind(Data,new_row)
View(Data)
```

```r
row.names(Data)[7]<-"седьмой"
View(Data)
```

```r
#Изменим название столбца

names(Data)[4]<-"rand_value"
View(Data)
```

## Получение сводных характеристик по прямоугольным данным, OLAP-средства в R

В R можно аналогично Excel выполнять агрегирование по данным фрейма, получая своеобразный механизм OLAP-средств. Для этого необходимо установить библиотеку rpivotTable

```r
install.packages('rpivotTable')
```

```r
#Подключаем библиотеку
library(rpivotTable)
```

```r
#Используем библиотеку на созданном ранее датафрейме
View(Data)
```

```r
#Расчитаем, сколько в сумме предметов каждого размера

rpivotTable::rpivotTable(Data,rows=c("size"), aggregatorName = "Sum",vals="count")
```

```r
#Рассчитаем количество предметов в разбивке по размеру
rpivotTable::rpivotTable(Data,rows=c("subject"), col=c("size"), aggregatorName = "Sum",vals="count")
```

```r
#Возможно разбиение с детализацией
rpivotTable::rpivotTable(Data,rows=c("size","subject"),  aggregatorName = "Median",vals="rand_value")
```
