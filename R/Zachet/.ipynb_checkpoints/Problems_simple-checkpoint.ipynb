**Первый Блок**

1. По двум из выбранных вами количественных признаков выполнить расчет мер центральной тендекции: среднего арифметического, гармонического, геометрического, медианы. 
2. Выполнить расчет значений 1, 3 квартилей для выбранных в п.1 признаков.
3. Построить по выбранным в п.1 признакам гистограмму частот, диаграмму - "ящик с усами", а также аппроксимацию плотности вероятности. Сделать выводы о скосах распределений.
5. Рассчитать по выбранным в п.1 показателям меры изменчивости данных: размах вариации, среднее квадратическое отклонение, дисперсию, коэффициент вариации, среднее линейное отклонение, интерквартильный размах.

**Второй Блок**

1. Оценить близость распределения выбранных количественных признаков к нормальному распределению, построив диаграмму квантиль-квантиль.
2. Выполнить проверку гипотез о нормальном распределении на основе теста Шапиро-Уилка.
3. Рассчитать линейный коэффициент корреляции между выбранными количественными показателями.

3. С использованием компонентного анализа провести кластеризацию количественных показателей датасета.
4. С использованием компонентного анализа провести кластеризацию объектов анализа датасета.
5. С использованием регрессионного анализа построить модель и провести прогнозирование выбранного показателя на тестовой выборке.
