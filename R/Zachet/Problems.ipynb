# Зачетное задание

В предложенном датасете выполнить следующие действия:
    
**Первый Блок**

1. По одному из выбранных вами количественных признаков выполнить расчет мер центральной тендекции: среднего арифметического, гармонического, геометрического, медианы. 
2. Выполнить расчет значений 1, 3 квартилей для выбранного в п.1 признака.
3. Построить по выбранному в п.1 признаку гистограмму частот, отметив на ней все рассчитанные меры центральной тенденции и квартили.
4. Выполнить группировку значений показателя, рассчитать среднее арифметическое, медиану и моду по сгруппированным данным, сравнить полученные значения с значениями, полученными по несгруппированным данным.
5. Рассчитать по выбранному в п.1 показателю меры изменчивости данных: размах вариации, среднее квадратическое отклонение, дисперсию, коэффициент вариации, среднее линейное отклонение, интерквартильный размах, коэффициент осцилляции, относительное линейное отклонение.
6. По выбранному в п.1 признаку построить коробку с усами, рассчитать показатели ассиметрии и эксцесса. Сделать выводы об ассиметрии распределения, степени остро или плосковершинности.

**Второй Блок**

1. Оценить близость распределения выбранных количественных признаков к нормальному распределению, построив диаграмму квантиль-квантиль.
2. Выполнить проверку гипотез о нормальном распределении на основе теста Шапиро-Уилка.
2. Рассчитать линейный коэффициент корреляции между выбранными количественными показателями, оценить его значимость, сделать выводы.
5. Выполнить расчет частного коэффициента корреляции между выбранными количественными показателями с исключением влияния других количественных признаков датасета.
6. Выполнить расчет ранговых коэффициентов корреляции Спирмена и Кендалла для двух выбранных количественных показателей, оценить их значимость.

**Третий Блок**
1. Выполнить нормирование всех количественных показателей приведя их к шкале от 0 до 100.
2. Построить корреляционную матрицу (с использованием цветовой гаммы) по всем нормированным показателям датасета.
3. Построить корреляционную матрицу (с использованием цветовой гаммы) по всем ненормированным показателям датасета.
4. Провести сравнение результатов п.2 и п.3.
5. С использованием компонентного анализа провести кластеризацию количественных показателей датасета.
6. С использованием компонентного анализа провести кластеризацию объектов анализа датасета.
7. Выполнить канонический корреляционный анализ, подобрав разбиение количественных показателей на две группы так, чтобы между сформированными по этим группам каноническим переменным наблюдалась максимальная корреляция.
8. С использованием выбранного метода (регрессионный анализ, дискриминантный анализ, метод опорных векторов, нейросети) построить модель и провести прогнозирование выбранного пока
